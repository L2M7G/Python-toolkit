```python
### Conditional Operators

import numpy as np

## Conditional Transformations - create a new column based on criteria

# np.where( condition, value if true, value if false) - assigns a value if condition is true, and another for when it is false

# ex: converting all temps to c if recorded in both c and f 

data_frame_name['new_column_name'] = np.where(
    data_frame_name['column_name_c_or_f'] == 'fahrenheit',      # condition based on a column denoting the measurement unit
    (data_frame_name['column_name_temp'] - 32) * 5/9,     # if condition is true, convert temp to celcius
    data_frame_name['column_name_temp']     # if condition is false, keep original temp value
)


## Binning - create descrete variabels by grouping numerical valies into bins (ex: age groups, income brackets, etc)

# np.select()  - assigns a value based on multiple conditions (vs np.where which only deals with True/False)
# with multiple conditions, follow first one that is true, so should order conditions from most specific to least specific (ex: above 90, above 80, above 70, etc.)
# assign conditions for n-1 categories and then use default for the nth category (ex: assign conditions for A, B, C, D and then use default to assign F for anything that doesn't meet the conditions for A-D)

# ex: assign a letter grade based on a numerical grade column

data_frame_name['new_column_name'] = np.select(
    [
        data_frame_name['column_name_grade'] >= 90,     # condition for A
        data_frame_name['column_name_grade'] >= 80,     # condition for B
        data_frame_name['column_name_grade'] >= 70,     # condition for C
        data_frame_name['column_name_grade'] >= 60      # condition for D
    ],
    [
        'A',    # value if condition for A is true
        'B',    # value if condition for B is true
        'C',    # value if condition for C is true
        'D'     # value if condition for D is true
    ],
    default = 'F'   # value if none of the conditions are true (ex: if grade is below 60, assign F)
)   

# can also assign conditions and outcomes before np.select() to make it easier to read

conditions = [
    data_frame_name['column_name_grade'] >= 90,     # condition for A
    data_frame_name['column_name_grade'] >= 80,     # condition for B
    data_frame_name['column_name_grade'] >= 70,     # condition for C
    data_frame_name['column_name_grade'] >= 60      # condition for D
]

outcomes = [
    'A',    # value if condition for A is true
    'B',    # value if condition for B is true
    'C',    # value if condition for C is true
    'D'     # value if condition for D is true
]

data_frame_name['new_column_name'] = np.select(conditions, outcomes, default = 'F')    # value if none of the conditions are true (ex: if grade is below 60, assign F)


# summary stats after conditional transformations 

new_data_frame_name = (data_frame_name
                       .groupby('transformed_column_name')
                       .agg(new_column_name_1 = ('column_name', 'mean')))



## Conditional Aggregation  

# look at a subset of the data that metts certain conditions
# think about the condition each value in the column needs to meet to be included in the summary statistic (ex: over a certain value, in a certain category, etc)

# .query() - added to the end after .agg() to filter the summary dataframe based on conditions 

summary_data_frame = (data_frame_name
                       .groupby('column_name')
                       .agg(new_column_name_1 = ('column_name', 'mean'))
                       .query('new_column_name_1 > 10')     # filter the summary dataframe to only include rows where the aggregated value is greater than 10
                       )


# Filter and aggregate the same column (.agg() with lambda) - ex: avg population by continent, but avg population for small population countries - kind of seeing it without the outliers

# lambda x: x[ x > 10] - use lambda to look at only x values in a column, and then look only at x values > 10

summary_data_frame = (data_frame_name
                          .groupby('continent')
                          .agg(
                                avg_pop = ('population', 'mean')
                                avg_pop_small = ('population', lambda x: x[x < 10_000_000].mean())    # get the average population for only the small population countries (population < 10 million) by using a lambda function to filter the population column to only include values less than 10 million before calculating the mean
                          )



# Filter by one column and aggregate a second (.apply() with lambda)
# apply() and lambda - when aggredations depends on conditions from another column (vs agg() is one column)

# ex: get the average gdp for countries with a population under a certain value - need to first filter for the population (1st column of interest), then average gdp (2nd column of interest)
# .apply(lambda x: pd.Series({
#        'new_column_name': x['column_name'].function(),
#        'new_column_name': x.loc[x['population'] < 10_000_000, 'column_name'].function()     # filter for population under 10 million and then calculate the summary statistic for the column of interest}))

summary_data_frame = (data_frame_name
                          .groupby('continent')
                          .apply(lambda x: pd.Series({
                                'avg_pop': x['population'].mean(),     # get the average population for each continent
                                'avg_pop_small': x.loc[x['population'] < 10_000_000, 'population'].mean(),    # get the average population for only the small population countries 
                                'avg_size': x['size'].mean(),     # can also calculate summary stats for other columns at the same time
                                'avg_gdp_small': x.loc[x['population'] < 10_000_000, 'gdp'].mean()    # filter for countries under 10 million people, and calculate the average gdp
                            }), include_groups = False)     # specify include_groups = False to not include the grouping column in the output dataframe, so just get the summary stats columns in the output dataframe
```
