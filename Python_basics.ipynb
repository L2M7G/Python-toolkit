```python
### Variables and Types

# Case-sensitive names 

age = 17
weight = 150
height = 1.70

print(age)

# ex: calculate BMI = w/h^2
bmi = weight/height **2
print(bmi)


# Variable types - type(variable_name)

print(type(age))

print(type(bmi))

# float - real numbers
# int - integer numbers
# str - string, text
# bool - True, False

x = "can double"
y = 'or single'

print(13 + 67)

('lou' + "isa")
```

```python
### Lists

list_name = [5, 7, 2, 8, 2, 7]
print(list_name)

# lists can have sub lists
sub_list = [[1, "a"],
            [2, "b"],
            [3, "c"]]

# can contain a mix or variable types

# Subsetting lists - select elements (index starts at 0)

countries = ["Canada", "USA", "Mexico", "Guatemala", "Belize", "Honduras", "Nicaragua", "Costa Rica", "Panama"]
print(countries[5])
print(countries[0])
print(countries[-1])   # -1 is useful to call the last element

# List Slicing - create a new list with selected elements [start:end] - (inclusive:exclusive)

print(countries[2:7])   
print(countries[:5])   # select the start of the list through 5
print(countries[5:])   # select the end of the list from 5

# to subset lists with sub lists [specify list][specify element]
print(sub_list[1][-1])


## Manipulating lists

# change list elements
list_name = [5, 7, 2, 8, 2, 7]
list_name[0] = 3   # change the first element in position 0 (5) to 3
print(list_name)
list_name[0:4] = [-3 , -1, 1, 3, 5]
print(list_name)

# Add elements
new_list = list_name + [2, 4, 6, 8]   # add additional list elements/ tack on another list
print(new_list)

# Delete elements
del new_list[0:2]
print(new_list)
```
