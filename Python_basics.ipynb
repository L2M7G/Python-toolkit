```python
### Dictionaries and Pandas

## Dictionaries
# Use to link two pieces of information
# indexed by keys (vs lists which are indexed by position)
# good to use when looking up data

countries = ["Canada", "USA", "Mexico"]
population = [40, 349, 130]

mex_index = countries.index("Mexico")
print(population[mex_index])

# instead use dictionaries to link countries to population
country_dict = {"Canada": 40, "USA": 349, "Mexico": 130}
print(country_dict["Mexico"])    # use Mexico as key to get population value

print(country_dict.keys())     # get the keys

# add additional pairs to the dictionary
country_dict["Guatemala"] = 17
print(country_dict)

# edit pairs
country_dict["Guatemala"] = 18
print(country_dict)

# delete pairs
del(country_dict)["Guatemala"]
print(country_dict)

# if the values are also a pair select an element - country_dict['guatemala']['capital]

# can make dictionary with lists
country = ["canada", "USA", "mexico"]
population = [40, 349, 130]
country_dict = {"Country": country, "Population": population}
print(country_dict)

## Pandas
# for data manipulation 
# store data in data frames

import pandas as pd

# can create a data frame from a dictionary - keys = column names, values = data for each column
country_dict = {"Country": ["Canada", "USA", "Mexico"], "Population": [40, 349, 130], "Capital": ["Ottawa", "Washington DC", "Mexico City"]}

country_df = pd.DataFrame(country_dict)
print(country_df)

# add an index column
country_df.index = ["CA", "US", "MX"]
print(country_df)

# data_name = pd.read_csv("file_name.csv", index_col=0) - read in data from a csv file, set the first column (0) as the index


# Index and select data - double brakets to return a data frame

# brakets - for selecting columns
print(country_df["Country"])    # select a column by name - returns a series

print(country_df[["Country"]])  # use double brackets [[]] to return a data frame
print(country_df[["Country", "Population"]])    # select multiple columns by name 

print(country_df[0:2])   # select rows by position (index starts at 0)

# loc - select rows and/or columnsby label ["Row Name"] or [:, "Column Name"] 
print(country_df.loc[["CA"]])
print(country_df.loc[["CA", "MX"], ["Population", "Capital"]])  # select specific rows and columns by label

# iloc - select rows and/or columns by position
print(country_df.iloc[:, 0:2])  # select columns by position
```

```python
### Logic, control FLow, and Filtering

import numpy as np

## Comparison operators 
# results in a boolean value (True or False)
# make comparisons between objects of the same type

print(4 == 4)
print(4 == 5)
print(4 <= 5)
print(4 >= 5)
print(4 != 5)

population = np.array([40, 349, 130])
print(population > 100)    # returns list of boolean values based on condition


## Boolean Operators - and, or, not
# combine multiple conditions

# and - both conditions must be true
print(True and True)    # True
print(True and False)   # False
print(False and False)  # False

x = 7
print(x > 5 and x < 10)   # check if x is between 5 and 10 -> True and True = True

# or - one or both conditions must be true
print(True or True)    # True
print(True or False)   # True
print(False or False)  # False

x = 7
print(x < 5 or x < 10)   # check if x is less than 5 or less than 10 -> False or True = True
print(x < 5 or x > 10)   # check if x is less than 5 or greater than 10 -> False or False = False

# not - negate a condition
print(not True)   # False
print(not False)  # True


# Numpy arrays
# does not work with combined conditions

# logical_and()
# logical_or()
# logical_not()

population = np.array([40, 349, 130])
print(np.logical_and(population > 100, population < 200))
print(np.logical_or(population < 50, population > 300))
print(np.logical_not(population > 100))

## if, elif, else

# if 

x = 10
if x % 2 == 0:     # %2 gives the remainder when x is divided by 2, if the remainder is 0 then x is even
    print("even")
else:
    print("odd")

x = 15
if x % 2 == 0:
    print("even")
else:
    print("odd")


## Filtering data frames
data = {"Country": ["Canada", "USA", "Mexico"], "Population": [40, 349, 130], "Capital": ["Ottawa", "Washington DC", "Mexico City"]}
country_df = pd.DataFrame(data)
country_df.index = ["CA", "US", "MX"]

country_df.loc[:, "Population"]
country_df["Population"] > 100
large_pop = country_df["Population"] > 100
print(large_pop)

# then use the boolean series large_pop to subset the data frame
print(country_df[large_pop])   # shows only the rows where the population is greater than 100 (columns for name, pop, and capital)

# one line code to do the same
country_df[country_df["Population"] > 100]

# Boolean operators
country_df[np.logical_and(country_df["Population"] > 100, country_df["Population"] < 200)]
```

```python
### Loops

## while loop - continues to repeat a bloack of code if the condition remains true (vs if statement which only runs once)
# good for repeating an action until a condition is met

error = 75
while error > 1:
    print("error:", error)
    error = error/3         # necessary to update the value of error, otherwise it will continue to print the same value of error and never end the loop

## for loop - repeat a block of code for a specified number of times, or once for each item in a list
# good for iterating through a list - for each element in the list, do something
countries = ["Canada", "USA", "Mexico", "Guatemala", "Belize", "Honduras", "Nicaragua", "Costa Rica", "Panama"]
for country in countries:
    print(country)

for index, country in enumerate(countries):   # enumerate gives both the index and the element in the list
    print("Number " + str(index + 1) + ": " + str(country))        # python index starts at 0, so index + 1 starts the printed numbers at 1


## Data Structures

# dictionaries   - use dictionary_name.items() to get both the key and value in the dictionary (method)
country_dict = {"Canada": 40, "USA": 349, "Mexico": 130, "Guatemala": 17}
for n, p in country_dict.items():    # .items() gives both the key and value in the dictionary
    print("The population of " + n + " is " + str(p) + " million")  

# numpy arrays     
miles_array = np.array([200, 345, 420, 180, 500])
gallon_array = np.array([10, 15, 14, 9, 20])
mpg = miles_array/gallon_array
for m in mpg:
    print(m)

# 2D numpy arrays  - use np.nditer(array_name) to iterate through each element in the 2D array (function)
miles_gallon = np.array([[200, 345, 420, 180, 500], 
                         [10, 15, 14, 9, 20]])
for mg in miles_gallon:    # this prints each row of the 2D array as a 1D array
    print(mg) 

for mg in np.nditer(miles_gallon):    # np.nditer() iterates through each element in the 2D array
    print(mg)                         # this prints each element in the 2D array on a new line


# data frame   - data_frame_name.iterrows() to get both the index and row of the data frame (method) (method)
country_df = pd.DataFrame({"Country": ["Canada", "USA", "Mexico", "Guatemala", "Belize", "Honduras", "Nicaragua", "Costa Rica", "Panama"], "Population": [40, 349, 130, 17, 0.4, 9.7, 6.5, 5.1, 4.3], "Capital": ["Ottawa", "Washington DC", "Mexico City", "Guatemala City", "Belmopan", "Tegucigalpa", "Managua", "San José", "Panama City"]})
country_df.index = ["CA", "US", "MX", "GT", "BZ", "HN", "NI", "CR", "PA"]

for index, row in country_df.iterrows():   # .iterrows() gives both the index and the row of the data frame
    print("Index: ", index)
    print("Row: ", row)

for index, row in country_df.iterrows():
    print(index + ": The population of " + row["Country"] + " is " + str(row["Population"]) + " million")

# add a column to the data frame with a for loop
for index, row in country_df.iterrows():
    country_df.loc[index, "Name Length"] = len(row["Country"])   # use .loc to add a value to a specific cell in the data frame, len() gives the length of the country name
    print(country_df)

# better to do that with apply
country_df["Name Length"] = country_df["Country"].apply(len)   # apply the len function to the "Country" column to create a new "Name Length" column
print(country_df)
```

```python
### Random Walks

import numpy as np
import matplotlib.pyplot as plt

## random generators

# np.random.seed(12345)   - set the seed for reproducibility
# np.random.rand()    - pseudo random number 

# ex: coin toss
np.random.seed(12345)
coin = np.random.randint(0,2)  # generates a random integer between 0 and 1 (0 or 1) to simulate a coin toss
print(coin)

if coin == 0: 
    print("Heads")
else:
    print("Tails")


# ex: dice roll
np.random.seed(12345)
dice = np.random.randint(1,7)
print(dice)

## Random Walk - next position depends on the current position and a random next step/change - stochastic process
# ex: stock price - next price depends on current price and random change in price

np.random.seed(12345)

# random steps - results of 20 coin tosses
result = []

for x in range(20):    # repeat the coin toss 20 times  
    coin = np.random.randint(0,2)  # 0 = heads, 1 = tails
    
    if coin == 0: 
        result.append("Heads")
    else:
        result.append("Tails")

print(result)

# random walk - number of tails over time
tails = [0]

for x in range(20):    # repeat the coin toss 20 times  
    coin = np.random.randint(0,2)  # 0 = heads, 1 = tails
    
    tails.append(tails[x] + coin)   # add the result of the coin toss to the previous number of tails to get the new number of tails

print(tails)

plt.plot(tails)
plt.show()

## Distribution
# ex: distribution of the final number of tails after 20 coin tosses

final_tails = []

for x in range(100000):   # repeat the random walk 100000 times - more times = more normal distribution
    tails = [0]
    
    for x in range(20):   
        coin = np.random.randint(0,2)  
        
        tails.append(tails[x] + coin)   
    
    final_tails.append(tails[-1])   # append the final number of tails (last element in the tails list) to the final_tails list

print(final_tails)
plt.hist(final_tails, bins = 20)
plt.show()

# histogram centered ~10 tails, which is ~50% of the 20 coin tosses (to be expected with a coin toss)
```
