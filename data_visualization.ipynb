```python
### Categorical and Quantiative Plots


## catplot() - same options as relplot (kind = "count" or "bar" or "box" or "point")


## Count Plot - shows count of each category

sns.countplot(x = 'column_name', data = data_frame_name)    

sns.catplot(x = 'column_name', data = data_frame_name, kind = "count")    # specify kind = "count" for count plot

# can create list of the categories to specify their order in the plot
order = ['category1', 'category2', 'category3']
sns.countplot(x = 'column_name', data = data_frame_name, order = order)

# order = list_name    - specify order of categories using list of category names


## Bar plot - shows summary statistic (ex: mean) for each category

sns.catplot(x = 'column_name', y = 'column_name', data = data_frame_name, kind = "bar")    # specify kind = "bar" for bar plot

# shows 95% CI by default

# errorbar = "sd"          - adjust to show the standard deviation instead of confidence interval

# errorbar = None          - remove confidence interval

# switch order of x and y to have horizontal bars instead of vertical bars


## Box Plot - info on median, quartiles, spread/skewness, and outliers

sns.catplot(x = 'column_name', y = 'column_name', data = data_frame_name, kind = "box")    # specify kind = "box" for box plot

# can change order as with other plots

# showfliers = False    - remove outliers from plot (points outside of whiskers)

# whis = 2.0            - adjust length of whiskers (default is 1.5 IQR)

# whis = [20, 80]       - specify which percentiles to show as whiskers (boxes show 25th and 75th)

# whis = [0, 100]       - show min and max values as whiskers (instead of IQR)


## Point Plot - shows meanof a variable for each category, with a line to show the CI (bar plot with points instead of bars) 

sns.catplot(x = 'column_name', y = 'column_name', data = data_frame_name, kind = "point")    # specify kind = "point" for point plot


# linestyle = "none"        - remove line between points

# estimator = np.median     - use median instead of mean for summary stat (must first import numpy as np, or from numpy import median)

# capsize = 0.3             - add caps to end of error bars (specify size of caps)

# errorbar = None           - remove error bars from plot
```
